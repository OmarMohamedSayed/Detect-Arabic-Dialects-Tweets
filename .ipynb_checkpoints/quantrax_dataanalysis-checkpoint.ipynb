```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("scoring.xlsx", names=["date","scores","company"])
```

```python
df.head()
```

## Remove Duplicated Rows

```python
#check if any row duplicated
df[df.duplicated(['date','scores','company'], keep=False)]
```

```python
#drop duplicated rows
df = df.drop_duplicates()
```

```python
df
```

## All Companies

```python
companies = df["company"].unique()
companies
```

## Total Scores For Each Company

```python
groupby_df = df.groupby(["company"]).sum()
```

```python
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(groupby_df.index,groupby_df.scores, color ='blue',
        width = 0.4)
 
plt.xlabel("Company Name")
plt.ylabel("No. of scores")
plt.title("Total Scores For Each Company")
plt.savefig('Total_Scores_For_Each_Company.jpg')
plt.show()
```

### Scores with Dates range 

```python
#convert dates to month period
per = df.date.dt.to_period("M") 
dates_and_scores_df = df.groupby(per).scores.sum()
dates_and_scores_df
```

```python
fig = plt.figure(figsize = (10, 5))
plt.plot(dates_and_scores_df.index.astype('str'),dates_and_scores_df.values,color='red')
plt.xlabel("Dates")
plt.ylabel("No. of scores")
plt.title("Scores in all dates")
plt.axhline(87000, color='b', linestyle='dashed')
plt.savefig('Scores_in_all_dates.jpg')
plt.show()
```

### All Scores above 87000 with Date period

```python
dates_and_scores_df[dates_and_scores_df.values>87000]
```

## Each Company scores in each date period

```python
def get_company_with_date(company):
    per = df.date.dt.to_period("M") 
    company_date_df = df[df["company"]==company].groupby([per]).scores.sum()
    return company_date_df
```

```python

figure, axes = plt.subplots(nrows=3, ncols=3)
count=0
for row in axes:
    for col in row:
        col.title.set_text(companies[count])
        company = get_company_with_date(companies[count])
        col.bar(company.index.astype(str),company.values)
        col.set(xlabel='dates', ylabel='scores')
        count=count+1
figure.set_size_inches(35, 15.5)

figure.tight_layout(pad=3.0)
plt.savefig('Each_Company_scores_in_each_date_period.jpg')
```
